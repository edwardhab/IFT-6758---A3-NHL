```python
import pandas as pd
import os
import json
```

```python
# Dossier contenant les fichiers JSON
json_folder_path = r'C:\Users\Admin\Documents\IFT6758-A3\IFT-6758---A3-NHL\nhl_data'

# Fichier CSV de sortie
output_csv_path = r'C:\Users\Admin\Documents\IFT6758-A3\IFT-6758---A3-NHL\output.csv'


data_list = [] 

# Parcourir tous les fichiers JSON dans le dossier
for filename in os.listdir(json_folder_path):
    if filename.endswith('.json'):
        json_file_path = os.path.join(json_folder_path, filename)
      
        with open(json_file_path, 'r') as f:
            data = json.load(f)
           
            data_list.append(data)


df = pd.DataFrame(data_list)

# Exporter les données en CSV
df.to_csv(output_csv_path, index=False)

print(f"Données JSON converties en CSV et sauvegardées dans {output_csv_path}")
```
